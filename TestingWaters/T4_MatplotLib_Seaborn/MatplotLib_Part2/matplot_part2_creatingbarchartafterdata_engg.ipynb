```python
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
# Using a style from matplot for more check https://www.dunderdata.com/blog/view-all-available-matplotlib-styles

plt.style.use("fivethirtyeight")
```

```python
# using read_csv method 
data = pd.read_csv('languages.csv')
# storing columns in variable i.e Responder_id,LanguagesWorkedWith
ids = data['Responder_id']

lang_responses = data['LanguagesWorkedWith']
```

```python
# Calling the Counter method to get the count
language_counter = Counter()
```

```python
# Simple loop for running over a variable and split is done as data contains ;
for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
# Taken 2 Lists
languages = []
popularity = []
```

```python
# Using a Simple Function and taking out most common counters
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
# Getting it reverse for making data in a order
languages.reverse()
popularity.reverse()
```

```python
# Use a barh function
plt.barh(languages, popularity)
```

```python
plt.barh(languages, popularity)
# Adding different attributes over this graph
plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```
